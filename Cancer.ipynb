```python
import statsmodels.api as sm
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cancer = sm.datasets.cancer.load_pandas().data
cancer.head()
```

```python
type(cancer)
```

```python
cancer.info()
```

```python
data = {
    'cancer': [1.0, 0.0, 3.0, 4.0, 3.0],
    'population': [445.0, 559.0, 677.0, 681.0, 746.0]
}

df = pd.DataFrame(data)
```

```python
df.plot(x='cancer', y='population', kind='line')
plt.xlabel('Cancer')
plt.ylabel('Population')
plt.title('Gráfico')
plt.show()
```

```python
df.plot(x='cancer', y='population', kind='bar')
plt.xlabel('Cancer')
plt.ylabel('Population')
plt.title('Gráfico de Barras')
plt.show()
```

```python
df.plot(x='cancer', y='population', kind='scatter')
plt.xlabel('Cancer')
plt.ylabel('Population')
plt.title('Gráfico de Dispersão')
plt.show()
```

```python
df.plot(x='cancer', y='population', kind='area')
plt.xlabel('Cancer')
plt.ylabel('Population')
plt.title('Gráfico de Área')
plt.show()
```

```python
df.plot(y='population', kind='pie', labels=df['cancer'], autopct='%1.1f%%')
plt.ylabel('Population')
plt.title('Gráfico de Pizza')
plt.show()
```

```python
df.plot(x='cancer', y='population', kind='line', marker='o')
plt.xlabel('Cancer')
plt.ylabel('Population')
plt.title('Gráfico de Linha com Marcadores')
plt.show()

```

```python
df.plot(kind='box')
plt.ylabel('Value')
plt.title('Gráfico de Caixas')
plt.show()
```
